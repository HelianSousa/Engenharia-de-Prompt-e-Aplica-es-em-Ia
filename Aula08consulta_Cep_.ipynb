#Consultas simples de APIs

### Consulta de Endereço via ViaCEP

Este script permite consultar informações de endereço utilizando a API pública do ViaCEP. Basta fornecer um CEP válido e ele retornará os dados formatados.

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API do ViaCEP para obter informações de endereço.

    Args:
        cep (str): O CEP a ser consultado (apenas números).

    Returns:
        str: Uma string formatada com as informações do endereço ou uma mensagem de erro.
    """
    # Remove qualquer caractere não numérico do CEP
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        return "Erro: CEP deve conter 8 dígitos."

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx)
        data = response.json()

        if 'erro' in data and data['erro']:
            return f"Erro: CEP '{cep}' não encontrado."
        else:
            endereco_formatado = (
                f"Endereço: {data.get('logradouro', 'N/A')}\n"
                f"Bairro: {data.get('bairro', 'N/A')}\n"
                f"Cidade: {data.get('localidade', 'N/A')}\n"
                f"Estado: {data.get('uf', 'N/A')}\n"
                f"CEP: {data.get('cep', 'N/A')}\n"
                f"DDD: {data.get('ddd', 'N/A')}\n"
                f"IBGE: {data.get('ibge', 'N/A')}"
            )
            return endereco_formatado

    except requests.exceptions.RequestException as e:
        return f"Erro na requisição: {e}"
    except json.JSONDecodeError:
        return "Erro: Não foi possível decodificar a resposta JSON da API."
    except Exception as e:
        return f"Ocorreu um erro inesperado: {e}"

# Exemplo de Uso:
#@title Digite o CEP para consulta
cep_para_consultar = "71693004" #@param {type:"string"}

resultado = consultar_cep(cep_para_consultar)
print(resultado)
```
